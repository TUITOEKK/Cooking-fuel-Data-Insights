```python
import pandas as pd

data = pd.read_csv("cooking_fuel.csv", parse_dates=["sale_date"])

# ── 1. Sort so gaps are calculated in chronological order
data = data.sort_values(["customer_id", "sale_date"])

# ── 2. Calculate day gaps between consecutive purchases per customer
data["day_gap"] = (
    data.groupby("customer_id")["sale_date"]
    .diff()                    # subtract previous date from current
    .dt.days                   # convert timedelta → integer days
)

# ── 3. Summarise per customer (drop NaN = first purchase has no gap)
summary = (
    data.dropna(subset=["day_gap"])
    .groupby(["customer_id", "first_name", "last_name"])["day_gap"]
    .agg(
        Purchases   = "count",   # number of gaps = purchases - 1
        Avg_Gap     = "mean",
        Median_Gap  = "median",
        Std_Gap     = "std",
        Min_Gap     = "min",
        Max_Gap     = "max"
    )
    .reset_index()
)

# ── 4. CV = Std / Mean  (how consistent the customer is)
summary["CV"] = summary["Std_Gap"] / summary["Avg_Gap"]

# ── 5. Keep only REGULAR customers (more than 1 purchase = has at least 1 gap)
regular = summary[summary["Purchases"] >= 2].copy()

print(f"REGULAR CUSTOMERS: {len(regular)}")
print(regular.sort_values("Avg_Gap").to_string(index=False))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import numpy as np

# ============================================================
# 1. IMPORT DATA
# ============================================================
data = pd.read_csv("cooking_fuel.csv", parse_dates=["sale_date"])
print(data.info())

# ============================================================
# 2. COUNT PURCHASES PER CUSTOMER
# ============================================================
purchase_counts = data.groupby("customer_id")["transaction_id"].count().reset_index()
purchase_counts.columns = ["customer_id", "Purchases"]

more_than_1 = purchase_counts[purchase_counts["Purchases"] > 1]
less_than_2  = purchase_counts[purchase_counts["Purchases"] < 2]

print("=" * 40)
print("     CUSTOMER PURCHASE SUMMARY")
print("=" * 40)
print(f"  Repeat customers (2+ purchases) : {len(more_than_1)}")
print(f"  One-time customers (1 purchase) : {len(less_than_2)}")
print("-" * 40)
print(f"  TOTAL unique customers          : {purchase_counts['customer_id'].nunique()}")
print("=" * 40)

# ============================================================
# 3. FILTER REPEAT CUSTOMERS & CALCULATE DAY GAPS
# ============================================================
repeat_customers = data[data["customer_id"].isin(more_than_1["customer_id"])].copy()
repeat_customers = repeat_customers.sort_values(["customer_id", "sale_date"])

repeat_customers["day_gap"] = (
    repeat_customers.groupby("customer_id")["sale_date"]
    .diff()
    .dt.days
)

# ============================================================
# 4. SUMMARISE GAPS PER CUSTOMER
# ============================================================
summary = (
    repeat_customers.dropna(subset=["day_gap"])
    .groupby(["customer_id", "first_name", "last_name"])["day_gap"]
    .agg(
        Num_Gaps   = "count",
        Avg_Gap    = "mean",
        Median_Gap = "median",
        Std_Gap    = "std",
        Min_Gap    = "min",
        Max_Gap    = "max"
    )
    .reset_index()
)

summary["Purchases"] = summary["Num_Gaps"] + 1
summary["CV"]        = summary["Std_Gap"] / summary["Avg_Gap"]

# ============================================================
# 5. CLASSIFY REGULAR vs IRREGULAR
# ============================================================
summary["Customer_Type"] = summary["CV"].apply(
    lambda x: "Regular" if x < 0.5 else "Irregular"
)

regular_count   = (summary["Customer_Type"] == "Regular").sum()
irregular_count = (summary["Customer_Type"] == "Irregular").sum()

print("=" * 45)
print("     CUSTOMER CLASSIFICATION SUMMARY")
print("=" * 45)
print(f"  Regular   customers (CV < 0.5)  : {regular_count}")
print(f"  Irregular customers (CV >= 0.5) : {irregular_count}")
print(f"  One-time  customers (1 purchase): {len(less_than_2)}")
print("-" * 45)
print(f"  TOTAL unique customers          : {len(purchase_counts)}")
print("=" * 45)

# ============================================================
# 6. PLOT CIRCLES FOR 29 IRREGULAR CUSTOMERS
# ============================================================
irregular     = summary[summary["Customer_Type"] == "Irregular"].copy()
irregular_ids = irregular["customer_id"].tolist()
irreg_data    = repeat_customers[repeat_customers["customer_id"].isin(irregular_ids)].copy()

# ── Grid layout: 10 rows x 3 cols = 30 slots for up to 29 customers
fig, axes = plt.subplots(nrows=10, ncols=3, figsize=(20, 40))
axes = axes.flatten()

for idx, cust_id in enumerate(irregular_ids):
    ax   = axes[idx]
    cust = irreg_data[irreg_data["customer_id"] == cust_id].sort_values("sale_date")
    dates = cust["sale_date"].values
    gaps  = cust["day_gap"].dropna().values
    name  = f"{cust['first_name'].iloc[0]} {cust['last_name'].iloc[0]}"

    cust_stats = irregular[irregular["customer_id"] == cust_id].iloc[0]

    # ── Timeline line
    ax.plot(dates, [1] * len(dates), color="gray", linewidth=1.2, zorder=2)

    # ── Purchase circles
    ax.scatter(dates, [1] * len(dates), s=120, color="tomato", 
               edgecolors="darkred", linewidth=0.8, zorder=3)

    # ── Gap labels between circles
    for i in range(1, len(dates)):
        mid_date = dates[i-1] + (dates[i] - dates[i-1]) / 2
        ax.annotate(
            f"{int(gaps[i-1])}d",
            xy=(mid_date, 1),
            xytext=(0, 14),
            textcoords="offset points",
            ha="center",
            fontsize=7.5,
            color="navy",
            fontweight="bold"
        )

    # ── Title with stats
    ax.set_title(
        f"{name}  |  Purchases: {int(cust_stats['Purchases'])}\n"
        f"Avg: {cust_stats['Avg_Gap']:.0f}d   "
        f"Median: {cust_stats['Median_Gap']:.0f}d   "
        f"CV: {cust_stats['CV']:.2f}",
        fontsize=8.5, fontweight="bold", pad=8
    )

    ax.xaxis.set_major_formatter(mdates.DateFormatter("%b %y"))
    ax.xaxis.set_major_locator(mdates.MonthLocator(interval=2))
    ax.set_yticks([])
    ax.tick_params(axis="x", labelsize=7, rotation=30)
    ax.set_ylim(0.75, 1.35)
    ax.grid(axis="x", linestyle="--", alpha=0.4)
    ax.spines[["left", "right", "top"]].set_visible(False)

# ── Hide unused subplots
for j in range(len(irregular_ids), len(axes)):
    axes[j].set_visible(False)

plt.suptitle(
    "IRREGULAR CUSTOMERS — Purchase Circles (CV ≥ 0.5)\n"
    "Each dot = purchase | Numbers = days between purchases",
    fontsize=14, fontweight="bold", y=1.005
)
plt.tight_layout()
plt.savefig("irregular_customers_circles.png", dpi=150, bbox_inches="tight")
plt.show()
print("✅ Chart saved as irregular_customers_circles.png")
```

```python
# ============================================================
# DETECT: Lapsed-then-Returned vs Truly Irregular
# ============================================================

def classify_irregular(cust_id):
    cust  = irreg_data[irreg_data["customer_id"] == cust_id].sort_values("sale_date")
    gaps  = cust["day_gap"].dropna().values
    
    if len(gaps) < 2:
        return "Insufficient data"
    
    avg_gap    = np.mean(gaps)
    median_gap = np.median(gaps)
    
    # A spike = any gap more than 2x the median
    spikes = gaps[gaps > 2 * median_gap]
    
    # Normal gaps = gaps within 2x median
    normal_gaps = gaps[gaps <= 2 * median_gap]
    
    spike_ratio = len(spikes) / len(gaps)  # proportion of spikes
    
    if spike_ratio <= 0.3:
        # Few spikes = mostly consistent with rare absence
        return "Lapsed-then-Returned"
    else:
        # Many spikes = genuinely unpredictable
        return "Truly Irregular"

# ── Apply classification
irregular["Sub_Type"] = irregular["customer_id"].apply(classify_irregular)

# ── Print results
lapsed  = (irregular["Sub_Type"] == "Lapsed-then-Returned").sum()
truly   = (irregular["Sub_Type"] == "Truly Irregular").sum()

print("=" * 50)
print("   IRREGULAR CUSTOMERS — DEEPER BREAKDOWN")
print("=" * 50)
print(f"  Lapsed-then-Returned  : {lapsed}")
print(f"  Truly Irregular       : {truly}")
print("-" * 50)
print(f"  TOTAL Irregular       : {len(irregular)}")
print("=" * 50)

print("\n── Lapsed-then-Returned Customers ──")
print(irregular[irregular["Sub_Type"] == "Lapsed-then-Returned"]
      [["first_name","last_name","Purchases","Avg_Gap","Median_Gap","CV"]]
      .to_string(index=False))

print("\n── Truly Irregular Customers ──")
print(irregular[irregular["Sub_Type"] == "Truly Irregular"]
      [["first_name","last_name","Purchases","Avg_Gap","Median_Gap","CV"]]
      .to_string(index=False))
```

```python
# ============================================================
# FIXED: Handle CV = NaN (customers with only 2 purchases = 1 gap)
# ============================================================

def classify_irregular_fixed(cust_id):
    cust   = irreg_data[irreg_data["customer_id"] == cust_id].sort_values("sale_date")
    gaps   = cust["day_gap"].dropna().values

    if len(gaps) < 2:
        # Only 1 gap — cannot calculate std/CV — label separately
        return "Single-Gap (2 purchases only)"

    median_gap  = np.median(gaps)
    spikes      = gaps[gaps > 2 * median_gap]
    spike_ratio = len(spikes) / len(gaps)

    if spike_ratio <= 0.3:
        return "Lapsed-then-Returned"
    else:
        return "Truly Irregular"

# ── Apply fixed classification
irregular["Sub_Type"] = irregular["customer_id"].apply(classify_irregular_fixed)

# ── Counts
lapsed      = (irregular["Sub_Type"] == "Lapsed-then-Returned").sum()
truly       = (irregular["Sub_Type"] == "Truly Irregular").sum()
single_gap  = (irregular["Sub_Type"] == "Single-Gap (2 purchases only)").sum()

print("=" * 55)
print("   IRREGULAR CUSTOMERS — FULL BREAKDOWN")
print("=" * 55)
print(f"  Lapsed-then-Returned          : {lapsed}")
print(f"  Truly Irregular               : {truly}")
print(f"  Single-Gap (2 purchases only) : {single_gap}")
print("-" * 55)
print(f"  TOTAL Irregular               : {len(irregular)}")
print("=" * 55)

# ── Print each group
for sub_type in ["Lapsed-then-Returned", "Truly Irregular", "Single-Gap (2 purchases only)"]:
    subset = irregular[irregular["Sub_Type"] == sub_type]
    print(f"\n── {sub_type} ({len(subset)} customers) ──")
    if len(subset) > 0:
        print(subset[["first_name","last_name","Purchases","Avg_Gap","Median_Gap","CV"]]
              .to_string(index=False))
    else:
        print("  None")
```

```python
# ============================================================
# INVESTIGATE: What drove Lapsed & Single-Gap customers back?
# ============================================================

lapsed_ids     = irregular[irregular["Sub_Type"] == "Lapsed-then-Returned"]["customer_id"].tolist()
single_gap_ids = irregular[irregular["Sub_Type"] == "Single-Gap (2 purchases only)"]["customer_id"].tolist()

def analyze_return_driver(cust_ids, label):
    cust_data = repeat_customers[repeat_customers["customer_id"].isin(cust_ids)].copy()
    cust_data = cust_data.sort_values(["customer_id", "sale_date"])

    # ── Tag each purchase as Before/During/After lapse
    results = []

    for cust_id in cust_ids:
        c     = cust_data[cust_data["customer_id"] == cust_id].copy()
        gaps  = c["day_gap"].dropna().values
        
        if len(gaps) == 0:
            continue

        median_gap = np.median(gaps)
        
        # Find the lapse row = where gap > 2x median
        c_rows = c.reset_index(drop=True)
        
        for i, row in c_rows.iterrows():
            gap_val = row["day_gap"]
            
            if pd.isna(gap_val):
                tag = "First Purchase"
            elif gap_val > 2 * median_gap:
                tag = "🔴 Return After Lapse"
            else:
                tag = "Normal Purchase"
            
            results.append({
                "customer_id"    : cust_id,
                "name"           : f"{row['first_name']} {row['last_name']}",
                "sale_date"      : row["sale_date"],
                "product_type"   : row["product_type"],
                "size"           : row["size"],
                "discount_amount": row["discount_amount"],
                "total_amount"   : row["total_amount"],
                "day_gap"        : gap_val,
                "purchase_tag"   : tag
            })

    df = pd.DataFrame(results)

    # ── Focus on RETURN purchases only
    returns = df[df["purchase_tag"] == "🔴 Return After Lapse"]

    print(f"\n{'='*60}")
    print(f"  {label}  —  {len(cust_ids)} customers")
    print(f"{'='*60}")

    # ── 1. Was there a discount on return?
    print("\n📌 1. DISCOUNT on Return Purchase?")
    discount_yes = (returns["discount_amount"] > 0).sum()
    discount_no  = (returns["discount_amount"] == 0).sum()
    print(f"     With discount    : {discount_yes} ({discount_yes/len(returns)*100:.1f}%)")
    print(f"     Without discount : {discount_no}  ({discount_no/len(returns)*100:.1f}%)")
    print(returns.groupby("discount_amount")
          .size().reset_index(name="count")
          .sort_values("discount_amount").to_string(index=False))

    # ── 2. Product type on return
    print("\n📌 2. PRODUCT TYPE on Return Purchase")
    print(returns["product_type"].value_counts().to_string())

    # ── 3. Size on return
    print("\n📌 3. SIZE on Return Purchase")
    print(returns["size"].value_counts().to_string())

    # ── 4. Compare: discount at return vs normal purchases
    print("\n📌 4. Avg Discount — Return vs Normal Purchases")
    print(df.groupby("purchase_tag")["discount_amount"]
          .agg(Count="count", Avg_Discount="mean", Max_Discount="max")
          .to_string())

    # ── 5. Full return detail per customer
    print(f"\n📌 5. RETURN DETAIL PER CUSTOMER")
    print(returns[["name","sale_date","product_type","size",
                   "discount_amount","total_amount","day_gap"]]
          .sort_values("day_gap", ascending=False)
          .to_string(index=False))

    return df

# ── Run for both groups
df_lapsed     = analyze_return_driver(lapsed_ids,     "LAPSED-THEN-RETURNED (14 customers)")
df_single_gap = analyze_return_driver(single_gap_ids, "SINGLE-GAP / 2 PURCHASES (29 customers)")

# ============================================================
# COMBINED SUMMARY — Discount vs No Discount
# ============================================================
print(f"\n{'='*60}")
print("  COMBINED SUMMARY — What Drove the Return?")
print(f"{'='*60}")

all_returns = pd.concat([
    df_lapsed[df_lapsed["purchase_tag"]     == "🔴 Return After Lapse"],
    df_single_gap[df_single_gap["purchase_tag"] == "🔴 Return After Lapse"]
])

print(f"\n  Total return events analysed : {len(all_returns)}")
print(f"\n  Discount drove return        : {(all_returns['discount_amount'] > 0).sum()}")
print(f"  No discount on return        : {(all_returns['discount_amount'] == 0).sum()}")
print(f"\n  Top product on return:")
print(all_returns["product_type"].value_counts().head(3).to_string())
print(f"\n  Top size on return:")
print(all_returns["size"].value_counts().head(3).to_string())
```

```python
# ============================================================
# BEFORE vs AFTER LAPSE — Product Type & Size Comparison
# ============================================================

fig, axes = plt.subplots(nrows=len(lapsed_ids), ncols=1, figsize=(14, len(lapsed_ids) * 4))

if len(lapsed_ids) == 1:
    axes = [axes]

for idx, cust_id in enumerate(lapsed_ids):
    ax   = axes[idx]
    c    = repeat_customers[repeat_customers["customer_id"] == cust_id].sort_values("sale_date").copy()
    gaps = c["day_gap"].fillna(0).values
    
    median_gap = np.median(c["day_gap"].dropna().values)
    name       = f"{c['first_name'].iloc[0]} {c['last_name'].iloc[0]}"

    # ── Find lapse point (gap > 2x median)
    lapse_idx = None
    for i, row in c.reset_index(drop=True).iterrows():
        if pd.notna(row["day_gap"]) and row["day_gap"] > 2 * median_gap:
            lapse_idx = i
            break

    if lapse_idx is None:
        continue

    before = c.iloc[:lapse_idx].copy()   # purchases before lapse
    after  = c.iloc[lapse_idx:].copy()   # purchases from return onwards

    # ── Build comparison table data
    before_products = before.groupby(["product_type", "size"]).size().reset_index(name="count")
    after_products  = after.groupby(["product_type", "size"]).size().reset_index(name="count")

    before_products["Period"] = "🔵 Before Lapse"
    after_products["Period"]  = "🟢 After Return"

    combined = pd.concat([before_products, after_products])
    combined["Product"] = combined["product_type"] + " | " + combined["size"]

    # ── Pivot for side by side
    pivot = combined.pivot_table(
        index   = "Product",
        columns = "Period",
        values  = "count",
        aggfunc = "sum"
    ).fillna(0).astype(int)

    # ── Ensure both columns exist
    for col in ["🔵 Before Lapse", "🟢 After Return"]:
        if col not in pivot.columns:
            pivot[col] = 0

    pivot = pivot[["🔵 Before Lapse", "🟢 After Return"]]
    pivot["Change"] = pivot["🟢 After Return"] - pivot["🔵 Before Lapse"]
    pivot["Change"] = pivot["Change"].apply(lambda x: f"▲ {x}" if x > 0 else (f"▼ {abs(x)}" if x < 0 else "─"))

    # ── Draw table
    ax.axis("off")
    lapse_date  = c.iloc[lapse_idx]["sale_date"].strftime("%b %Y")
    lapse_days  = int(c.iloc[lapse_idx]["day_gap"])

    table_data   = [[idx_val] + list(row) for idx_val, row in zip(pivot.index, pivot.values)]
    col_labels   = ["Product | Size", "🔵 Before Lapse", "🟢 After Return", "Change"]
    
    table = ax.table(
        cellText    = table_data,
        colLabels   = col_labels,
        cellLoc     = "center",
        loc         = "center"
    )
    table.auto_set_font_size(False)
    table.set_fontsize(9)
    table.scale(1.2, 1.6)

    # ── Color header
    for j in range(len(col_labels)):
        table[0, j].set_facecolor("#2c3e50")
        table[0, j].set_text_props(color="white", fontweight="bold")

    # ── Color rows by period change
    for i in range(1, len(table_data) + 1):
        change_val = pivot["Change"].iloc[i-1]
        for j in range(len(col_labels)):
            if "▲" in str(change_val):
                table[i, j].set_facecolor("#d5f5e3")   # green = increased after return
            elif "▼" in str(change_val):
                table[i, j].set_facecolor("#fde8e8")   # red = decreased after return
            else:
                table[i, j].set_facecolor("#f8f9fa")   # grey = no change

    ax.set_title(
        f"{name}  |  Lapse: {lapse_days} days ({lapse_date})  |  "
        f"Before: {len(before)} purchases   After: {len(after)} purchases",
        fontsize=10, fontweight="bold", pad=12, loc="left"
    )

plt.suptitle(
    "LAPSED-THEN-RETURNED CUSTOMERS\nProduct Type & Size — Before vs After Lapse",
    fontsize=13, fontweight="bold", y=1.01
)
plt.tight_layout()
plt.savefig("lapsed_before_after_products.png", dpi=150, bbox_inches="tight")
plt.show()
print("✅ Saved as lapsed_before_after_products.png")
```

```python
# ============================================================
# 1. REVENUE AFTER DISCOUNT — Net Revenue per Product
# ============================================================

revenue = repeat_customers.groupby("product_type").agg(
    Total_Purchases  = ("transaction_id", "count"),
    Gross_Revenue    = ("total_amount", "sum"),
    Total_Discount   = ("discount_amount", "sum"),
).reset_index()

revenue["Net_Revenue"]        = revenue["Gross_Revenue"] - revenue["Total_Discount"]
revenue["Avg_Net_Per_Sale"]   = (revenue["Net_Revenue"] / revenue["Total_Purchases"]).round(2)
revenue["Discount_Pct"]       = (revenue["Total_Discount"] / revenue["Gross_Revenue"] * 100).round(1)

print("=" * 85)
print("  1. NET REVENUE AFTER DISCOUNT PER PRODUCT")
print("=" * 85)
print(revenue[["product_type","Total_Purchases","Gross_Revenue",
               "Total_Discount","Net_Revenue","Avg_Net_Per_Sale","Discount_Pct"]]
      .sort_values("Net_Revenue", ascending=False)
      .to_string(index=False))

# ============================================================
# 2. AMERIGAS CUSTOMER LOYALTY vs DISCOUNT HUNTING
# ============================================================

# ── Get all AmeriGas buyers
amerigas_buyers = repeat_customers[repeat_customers["product_type"] == "AmeriGas"].copy()

# ── Per customer: how many times they bought AmeriGas vs other products
customer_product = repeat_customers.groupby(["customer_id","product_type"]).size().reset_index(name="count")

amerigas_ids = amerigas_buyers["customer_id"].unique()

loyalty_results = []

for cust_id in amerigas_ids:
    cust_purchases = customer_product[customer_product["customer_id"] == cust_id]
    total          = cust_purchases["count"].sum()
    amerigas_count = cust_purchases[cust_purchases["product_type"] == "AmeriGas"]["count"].sum()
    loyalty_pct    = amerigas_count / total * 100

    # Check if their AmeriGas purchases had discounts
    cust_amerigas  = amerigas_buyers[amerigas_buyers["customer_id"] == cust_id]
    disc_purchases = (cust_amerigas["discount_amount"] > 0).sum()
    total_amerigas = len(cust_amerigas)
    disc_rate      = disc_purchases / total_amerigas * 100

    loyalty_results.append({
        "customer_id"       : cust_id,
        "AmeriGas_Purchases": amerigas_count,
        "Total_Purchases"   : total,
        "Loyalty_Pct"       : round(loyalty_pct, 1),
        "Discounted_Visits" : disc_purchases,
        "Discount_Rate_Pct" : round(disc_rate, 1)
    })

loyalty_df = pd.DataFrame(loyalty_results)

# ── Classify each AmeriGas customer
def classify_buyer(row):
    if row["Loyalty_Pct"] >= 70 and row["Discount_Rate_Pct"] < 30:
        return "💚 Loyal — Brand Preference"
    elif row["Loyalty_Pct"] >= 70 and row["Discount_Rate_Pct"] >= 30:
        return "🟡 Loyal — Possibly Discount Driven"
    elif row["Loyalty_Pct"] < 70 and row["Discount_Rate_Pct"] >= 30:
        return "🔴 Discount Hunter — Not Loyal"
    else:
        return "🔵 Occasional Buyer"

loyalty_df["Buyer_Type"] = loyalty_df.apply(classify_buyer, axis=1)

# ── Summary counts
print("\n" + "=" * 55)
print("  2. AMERIGAS — LOYALTY vs DISCOUNT HUNTING")
print("=" * 55)
print(loyalty_df["Buyer_Type"].value_counts().to_string())

# ── Averages per buyer type
print("\n── Average Stats per Buyer Type ──")
print(loyalty_df.groupby("Buyer_Type")[["Loyalty_Pct","Discount_Rate_Pct","AmeriGas_Purchases"]]
      .mean().round(1).to_string())

# ── Full detail
print("\n── Full Customer Detail ──")
print(loyalty_df.sort_values("Loyalty_Pct", ascending=False).to_string(index=False))
```

```python
# ============================================================
# REVENUE ANALYSIS BY PRODUCT + SIZE
# ============================================================

# ── 1. See what sizes exist per product
print("=" * 55)
print("  SIZES AVAILABLE PER PRODUCT")
print("=" * 55)
size_matrix = repeat_customers.groupby(["product_type","size"]).size().unstack(fill_value=0)
print(size_matrix.to_string())

# ── 2. Revenue breakdown by product AND size
print("\n" + "=" * 90)
print("  REVENUE BY PRODUCT + SIZE")
print("=" * 90)
size_revenue = repeat_customers.groupby(["product_type","size"]).agg(
    Purchases      = ("transaction_id", "count"),
    Gross_Revenue  = ("total_amount", "sum"),
    Total_Discount = ("discount_amount", "sum"),
    Avg_Unit_Price = ("unit_price", "mean"),
    Avg_Quantity   = ("quantity", "mean")
).reset_index()

size_revenue["Net_Revenue"]      = size_revenue["Gross_Revenue"] - size_revenue["Total_Discount"]
size_revenue["Avg_Net_Per_Sale"] = (size_revenue["Net_Revenue"] / size_revenue["Purchases"]).round(2)
size_revenue["Discount_Pct"]     = (size_revenue["Total_Discount"] / size_revenue["Gross_Revenue"] * 100).round(1)

print(size_revenue[["product_type","size","Purchases","Avg_Unit_Price",
                     "Avg_Quantity","Gross_Revenue","Total_Discount",
                     "Net_Revenue","Avg_Net_Per_Sale","Discount_Pct"]]
      .sort_values(["product_type","size"])
      .to_string(index=False))

# ── 3. Compare same size across products — APPLES TO APPLES
print("\n" + "=" * 90)
print("  SAME SIZE COMPARISON ACROSS PRODUCTS — Fair Comparison")
print("=" * 90)

sizes = repeat_customers["size"].unique()

for size in sorted(sizes):
    subset = size_revenue[size_revenue["size"] == size].sort_values("Avg_Net_Per_Sale", ascending=False)
    if len(subset) > 1:
        print(f"\n── Size: {size} ──")
        print(subset[["product_type","Purchases","Avg_Unit_Price",
                       "Avg_Net_Per_Sale","Discount_Pct"]]
              .to_string(index=False))

# ── 4. Avg unit price per size — is pricing consistent?
print("\n" + "=" * 60)
print("  AVG UNIT PRICE PER SIZE (across all products)")
print("=" * 60)
print(repeat_customers.groupby("size")["unit_price"]
      .agg(Count="count", Avg_Price="mean", Min_Price="min", Max_Price="max")
      .sort_values("Avg_Price", ascending=False)
      .to_string())
```

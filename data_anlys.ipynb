```python
data.rename(columns = {"size":"size(lb)"}, inplace = True)
```

```python
data["size(lb)"] = data["size(lb)"].str.replace("lb","",regex = False).astype(float)
```

```python
data["sale_date"] = pd.to_datetime(data["sale_date"])
```

```python
print(data["sale_date"].dtype)
```

```python
data["name"] = (data["first_name"].fillna("") + " " + data["last_name"].fillna("")).str.strip()
```

```python
data.head()
```

```python
data = data.drop(columns = ["first_name","last_name"])
```

```python
col = data.pop("name")
data.insert(2, "name" , col)
```

```python
data = data.set_index("sale_date")
```

```python
monthly = data.resample("M").size()
```

```python
monthly_orders = data.resample("M").size()
```

```python
import matplotlib.pyplot as plt

monthly_orders.plot()
plt.title("Monthly Orders")
plt.xlabel("Month")
plt.ylabel("Number of Orders")
plt.show()
```

```python
monthly_discount = data[data["discount_amount"] > 0].resample("M").size()
```

```python
import matplotlib.pyplot as plt

monthly_orders.plot()
monthly_discount.plot()

plt.title("Monthly Orders vs Discount Orders")
plt.xlabel("Month")
plt.ylabel("Number of Orders")
plt.legend(["Total Orders", "Discount Orders"])
plt.show()
```

```python
import pandas as pd
data = pd.read_csv("cooking_fuel.csv")
```

```python
# ===== STEP 1: PREP =====
data['sale_date'] = pd.to_datetime(data['sale_date'])
data = data.sort_values(['customer_id', 'sale_date']).reset_index(drop=True)
data['day_gap'] = data.groupby('customer_id')['sale_date'].diff().dt.days
```

```python
# ===== STEP 2: BUILD GAP PROFILE PER CUSTOMER =====
gap_profile = (data[data['day_gap'].notna()]
               .groupby('customer_id')['day_gap']
               .agg(
                   num_gaps   = 'count',
                   avg_gap    = 'mean',
                   median_gap = 'median',
                   std_gap    = 'std',
                   min_gap    = 'min',
                   max_gap    = 'max'
               ).round(1).reset_index())

# CV = how consistent is their cycle
# Low  CV → regular (buys on a steady rhythm)
# High CV → irregular (unpredictable)
gap_profile['CV'] = (gap_profile['std_gap'] / gap_profile['avg_gap']).round(2)

# Attach name
names = data.groupby('customer_id')['first_name'].first().reset_index()
gap_profile = gap_profile.merge(names, on='customer_id')
```

```python
# ===== STEP 3: CLASSIFY EACH CUSTOMER =====
def classify(row):
    if row['num_gaps'] < 2:
        return 'too few purchases'
    if row['CV'] < 0.3:
        return 'regular'        # steady clock, low variation
    elif row['CV'] < 0.6:
        return 'semi-regular'   # some variation but pattern exists
    else:
        return 'irregular'      # unpredictable

gap_profile['circle_type'] = gap_profile.apply(classify, axis=1)

print("=" * 40)
print("   CUSTOMER CIRCLE CLASSIFICATION")
print("=" * 40)
print(gap_profile['circle_type'].value_counts())
```

```python
# ===== FIX THE CLASSIFIER =====
# Original wrong logic:
# if row['num_gaps'] < 2:
#     return 'too few gaps'   ← this discards customers with 2 purchases (1 gap)

# Fixed logic — only exclude customers with 0 gaps (1 purchase)
def classify(row):
    if row['num_gaps'] < 1:                  # 0 gaps = 1 purchase = truly no circle
        return 'too few purchases'
    if pd.isna(row['CV']):                   # only 1 gap → std is NaN → CV is NaN
        # can't measure consistency with 1 gap
        # classify by gap size only
        if row['avg_gap'] < 45:
            return 'regular'
        elif row['avg_gap'] < 90:
            return 'semi-regular'
        else:
            return 'irregular'
    # 2+ gaps → CV is reliable → use it normally
    if row['CV'] < 0.3:
        return 'regular'
    elif row['CV'] < 0.6:
        return 'semi-regular'
    else:
        return 'irregular'

gap_profile['circle_type'] = gap_profile.apply(classify, axis=1)

# ===== VERIFY COUNTS NOW =====
print("=== Classification Counts ===")
print(gap_profile['circle_type'].value_counts())
print(f"\nTotal in gap_profile         : {len(gap_profile)}")
print(f"1 purchase (outside profile) : 26")
print(f"Grand total                  : {len(gap_profile) + 26}")
print(f"Matches data.customer_id     : {len(gap_profile) + 26 == data['customer_id'].nunique()}")
```

```python
# customers with only 1 purchase
one_purchase = (data.groupby('customer_id')
                .size()
                .reset_index(name='total_purchases')
                .query('total_purchases == 1'))

# get their details + gap to end of dataset
last_date_in_data = data['sale_date'].max()

one_purchase_details = one_purchase.merge(
    data[['customer_id','first_name','sale_date']],
    on='customer_id'
)

one_purchase_details['gap_to_end'] = (last_date_in_data - 
                                       pd.to_datetime(one_purchase_details['sale_date'])).dt.days

print(f"Last date in dataset: {last_date_in_data.date()}")
print(f"Total customers with 1 purchase: {len(one_purchase_details)}")
print()
print(one_purchase_details[['customer_id','first_name','sale_date',
                             'gap_to_end']].sort_values('gap_to_end', ascending=False)
                                           .to_string(index=False))
```

```python
# who is in the 291 but NOT in gap_profile and NOT in one_purchase?
all_customers      = set(data['customer_id'].unique())           # 291
in_gap_profile     = set(gap_profile['customer_id'].unique())    # 242
in_one_purchase    = set(one_purchase_details['customer_id'].unique())  # 26

accounted_for      = in_gap_profile | in_one_purchase            # 242 + 26 = 268
unaccounted        = all_customers - accounted_for               # 23 missing

print(f"Total unique customers : {len(all_customers)}")
print(f"In gap_profile         : {len(in_gap_profile)}")
print(f"1 purchase customers   : {len(in_one_purchase)}")
print(f"Accounted for          : {len(accounted_for)}")
print(f"Still unaccounted      : {len(unaccounted)}")

# inspect them
unaccounted_df = data[data['customer_id'].isin(unaccounted)]
print("\n=== Unaccounted Customers ===")
print(unaccounted_df.groupby('customer_id')
      .agg(
          first_name      = ('first_name', 'first'),
          total_purchases = ('transaction_id', 'count'),
          first_date      = ('sale_date', 'min'),
          last_date       = ('sale_date', 'max')
      ).reset_index().to_string(index=False))
```

```python
# ===== STEP 5: VISUALIZE ALL THREE GROUPS =====
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

regular     = gap_profile[gap_profile['circle_type'] == 'regular']
semi        = gap_profile[gap_profile['circle_type'] == 'semi-regular']
irregular   = gap_profile[gap_profile['circle_type'] == 'irregular']

fig = plt.figure(figsize=(16, 12))
gs  = gridspec.GridSpec(2, 3, figure=fig)

# ── Row 1: CV distribution per group ──────────────────────────────────
ax1 = fig.add_subplot(gs[0, :])   # full width top row
colors_map = {'regular':'green', 'semi-regular':'gold', 'irregular':'tomato'}
for ctype, grp in gap_profile[gap_profile['circle_type'] != 'too few purchases'].groupby('circle_type'):
    ax1.scatter(grp['avg_gap'], grp['CV'],
                label=ctype, color=colors_map[ctype],
                s=80, edgecolors='black', alpha=0.8)

ax1.axhline(0.3, color='green',  linestyle='--', alpha=0.6, label='CV=0.3 (regular boundary)')
ax1.axhline(0.6, color='tomato', linestyle='--', alpha=0.6, label='CV=0.6 (irregular boundary)')
ax1.set_title('All Customers — Avg Gap vs Consistency (CV)\nLower CV = More Regular', fontsize=12)
ax1.set_xlabel('Average Gap (days)')
ax1.set_ylabel('CV (Coefficient of Variation)')
ax1.legend()
ax1.grid(True, alpha=0.3)

# ── Row 2: Gap distributions per group ────────────────────────────────
for ax, grp, title, color in zip(
    [fig.add_subplot(gs[1, 0]),
     fig.add_subplot(gs[1, 1]),
     fig.add_subplot(gs[1, 2])],
    [regular, semi, irregular],
    ['Regular', 'Semi-Regular', 'Irregular'],
    ['green', 'gold', 'tomato']
):
    ax.hist(grp['avg_gap'], bins=15, color=color, edgecolor='black', alpha=0.8)
    ax.set_title(f'{title}\n({len(grp)} customers)', fontsize=11)
    ax.set_xlabel('Avg Gap (days)')
    ax.set_ylabel('Count')
    ax.grid(True, alpha=0.3)

plt.suptitle('Customer Purchase Circle Analysis', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
# ===== STEP 6: SHOW INDIVIDUAL RHYTHM FOR EACH GROUP =====
# Pick 2 sample customers from each group and plot their actual gaps
import numpy as np
def plot_sample_gaps(group_df, title, color, n=2):
    samples = group_df.head(n)['customer_id'].values
    fig, axes = plt.subplots(1, n, figsize=(7 * n, 4))
    if n == 1: axes = [axes]

    for ax, cid in zip(axes, samples):
        cust = data[(data['customer_id'] == cid) & data['day_gap'].notna()]
        name = cust['first_name'].iloc[0]
        gaps = cust['day_gap'].values

        ax.bar(range(len(gaps)), gaps, color=color, edgecolor='black', alpha=0.85)
        ax.axhline(gaps.mean(),   color='black', linestyle='--', label=f'avg={gaps.mean():.0f}d')
        ax.axhline(np.median(gaps), color='blue', linestyle=':', label=f'median={np.median(gaps):.0f}d')
        ax.set_title(f'{title}: {name} (ID:{cid})')
        ax.set_xlabel('Purchase Cycle Number')
        ax.set_ylabel('Days Since Last Purchase')
        ax.legend()
        ax.grid(True, alpha=0.3)

    plt.suptitle(f'{title} — Individual Gap Rhythms', fontweight='bold')
    plt.tight_layout()
    plt.show()

plot_sample_gaps(regular,   'REGULAR',      'green')
plot_sample_gaps(semi,      'SEMI-REGULAR', 'gold')
plot_sample_gaps(irregular, 'IRREGULAR',    'tomato')
```

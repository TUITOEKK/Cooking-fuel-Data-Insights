1.Descriptive analysis  

We will start with identifying trends of monthly sales

```python
data["transaction_id"].nunique()
```

I first checked the `transaction_id` column to confirm if it can represent sales. The column has 1632 unique values, which matches the total number of rows in the dataset. This shows each row represents a unique transaction. Therefore, `transaction_id` is a valid measure of total sales.


```python
data["month"] = data["sale_date"].dt.strftime("%b")
```

```python
data.head(3)
```

```python
month_order = ["Jan","Feb","Mar","Apr","May","Jun",
         "Jul","Aug","Sep","Oct","Nov","Dec"]
```

```python
sales = data.groupby("month")["transaction_id"].count().reindex(month_order)
print(sales)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 4))
plt.plot(sales.index, sales.values, marker="o", color="blue", linewidth=2)


plt.title("Monthly Sales")
plt.xlabel("Month")
plt.ylabel("Number of Sales")


plt.xticks(rotation=45)  
plt.tight_layout()        
plt.show()
```

From the chart we can see that february records the highest number of sales which is 174 while Novermber has got the least number of sales of 94.Sales recover in April and remain fairly stable through July, ranging between 144 and 155. This period shows consistent performance with moderate fluctuations.
From August, sales begin to decline. September and October continue at lower levels compared to mid year.
November shows the lowest sales, around 94. This is a significant drop and stands out as the weakest month while December shows a slight recovery but remains low compared to most months.

This pattern suggests strong early demand, stable middle performance, and weak sales toward the end of the year.

```python
state_sales = data.groupby("state")["transaction_id"].count().sort_values(ascending = False)
state_sales.head(10)
```

```python
state_sales.tail(10)
```

```python
top_10 =state_sales.head(10)
bottom_10 = state_sales.tail(10)
```

```python
# Sort values
top_10 = top_10.sort_values(ascending=True)
bottom_10 = bottom_10.sort_values(ascending=True)

# Create side by side charts
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

# ── Top 10 ──
bars1 = ax1.barh(top_10.index, top_10.values, color="#0DD13A")
ax1.set_title("Top 10 States by Sales", fontweight="bold")
ax1.set_xlabel("Number of Transactions")

for bar in bars1:
    width = bar.get_width()
    ax1.text(width, bar.get_y() + bar.get_height() / 2,
             f" {int(width)}",
             va="center", ha="left",
             fontsize=9, color="black")

# ── Bottom 10 ──
bars2 = ax2.barh(bottom_10.index, bottom_10.values, color="salmon")
ax2.set_title("Bottom 10 States by Sales", fontweight="bold")
ax2.set_xlabel("Number of Transactions")

for bar in bars2:
    width = bar.get_width()
    ax2.text(width, bar.get_y() + bar.get_height() / 2,
             f" {int(width)}",
             va="center", ha="left",
             fontsize=9, color="black")

plt.tight_layout()
plt.show()
```

Top performing states show strong transaction volumes. Georgia leads with 86 transactions. North Carolina follows with 84. Tennessee comes next with 78. These states stand out as the main drivers of sales.
The rest of the top group stays within a close range, from about 62 to 69. This shows stable performance among the top 10, with no extreme gaps after the top three.Bottom performing states show much lower activity. Colorado leads this group with 47, which is still far below the top states. Pennsylvania and Wisconsin follow with 44 and 43.The weakest state is Virginia with only 25 transactions. This is less than one third of Georgia’s volume, showing a large performance gap.
This pattern shows uneven sales distribution, where a few states generate most of the activity while others contribute much less.

In terms of revenue Blue Rhino generates the highest total revenue at 240406.47, followed by Suburban Propane. Ferrellgas has the lowest revenue even though it has one of the higher transaction counts.

2. Diagnostic Analysis  
In this I'm going to focuses on understanding the reasons behind the patterns seen in the data. It goes beyond describing what happened and works to explain why it happened. The goal is to connect changes in key metrics such as sales, customers, discounts, or revenue to possible causes like pricing, product performance, or customer behavior.
I will start by analyzing what drives sales, since I observed that sales fluctuate as the months increase. I will break down the data step by step to identify the key factors behind these changes. This includes looking at trends in discounts to understand what is influencing sales over time. The goal is to clearly identify the reasons for the fluctuations and determine which factors have the strongest impact on sales.


```python
monthly_sales = data.groupby("month")["transaction_id"].count().reindex(month_order)
print(monthly_sales)
```

```python
discount_count = data[data["discount_amount"] > 0]
```

```python
monthly_discount_count= discount_count.groupby("month")["discount_amount"].count().reindex(month_order)
print(monthly_discount_count)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 4))
monthly_sales.plot()
monthly_discount_count.plot()

plt.title("Monthly Orders vs Discount Orders")
plt.xlabel("Month")
plt.ylabel("Number of Orders")
plt.legend(["Total Orders", "Discount Orders"])
plt.show()
```

```python
discount_orders = data[data["discount_amount"] > 0]
no_discount_orders = data[data["discount_amount"] == 0]
```

```python
avg_discount_orders_quantity = discount_orders["quantity"].mean()
avg_no_discount_orders_quantity = no_discount_orders["quantity"].mean()
```

```python
labels = ["Discount Orders", "No Discount Orders"]
values = [avg_discount_orders_quantity, avg_no_discount_orders_quantity]

# Plot
plt.figure()
plt.bar(labels, values)

# Labels and title
plt.xlabel("Order Type")
plt.ylabel("Average Quantity")
plt.title("Average Order Quantity: Discount vs No Discount")

plt.show()
```

```python
avg_discount_orders_revenue = discount_orders["total_amount"].mean()
avg_no_discount_orders_revenue = no_discount_orders["total_amount"].mean()
```

```python
plt.figure()
plt.bar(
    ["Discount Orders", "No Discount Orders"],
    [avg_discount_orders_revenue, avg_no_discount_orders_revenue]
)

plt.xlabel("Order Type")
plt.ylabel("Average Revenue")
plt.title("Average Revenue per Order")

plt.show()
```

```python
print(data.groupby("customer_id")["discount_amount"].sum().describe())
```

```python
import matplotlib.pyplot as plt

# Create a copy
df = data.sort_values("sale_date").copy()

# First purchase per customer
first_purchase = df.groupby("customer_id").first().reset_index()

# Extract month
first_purchase["month"] = first_purchase["sale_date"].dt.strftime("%b")

# New customers per month
monthly_new_customers = first_purchase.groupby("month")["customer_id"].nunique()

# New customers with discount
monthly_new_discount_customers = first_purchase[first_purchase["discount_amount"] > 0].groupby("month")["customer_id"].nunique()

monthly_new_customers = monthly_new_customers.reindex(month_order)
monthly_new_discount_customers = monthly_new_discount_customers.reindex(month_order)

# Plot
plt.figure(figsize=(8, 4))

monthly_new_customers.plot()
monthly_new_discount_customers.plot()

plt.title("New Customers vs Discount New Customers")
plt.xlabel("Month")
plt.ylabel("Number of Customers")
plt.legend(["New Customers", "Discount New Customers"])

plt.show()
```

```python
monthly_table = pd.DataFrame({
    "New Customers": monthly_new_customers,
    "Discount New Customers": monthly_new_discount_customers
})

monthly_table = monthly_table.fillna(0).astype(int)

print(monthly_table)
```

```python
data = data.drop(columns = ["first_name","last_name"])
```

```python
data = data.sort_values(['customer_id', 'sale_date'])
```

```python
data['prev_date'] = data.groupby('customer_id')['sale_date'].shift(1)

data['gap_days'] = (data['sale_date'] - data['prev_date']).dt.days
```

```python
data.tail(20)
```

```python

# analysis date (dataset snapshot date)
analysis_date = data['sale_date'].max()

# build customer-level table
customer_summary = data.groupby('customer_id').agg(
    customer_name=('name', 'first'),
    gaps=('gap_days', lambda x: list(x.dropna())),   # list like [20, 24, 45, 60]
    mean_gap=('gap_days', 'mean'),
    median_gap=('gap_days', 'median'),
    last_purchase=('sale_date', 'max')
).reset_index()

# compute current gap
customer_summary['current_gap'] = (
    analysis_date - customer_summary['last_purchase']
).dt.days

# nice ordering
customer_summary = customer_summary[
    ['customer_id', 'customer_name', 'gaps', 'mean_gap', 'median_gap', 'current_gap']
]

customer_summary
```
